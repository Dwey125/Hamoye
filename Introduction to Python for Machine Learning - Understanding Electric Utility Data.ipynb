```python
#Question 6
df.isna().sum()
```

```python
percent_missing = df.isnull().sum() * 100 / len(df)
percent_missing * 100
```

```python
#Question 7
#Calculating the modal for the fuel_unit column i.e The most common value in the fuel_unit column
#This helps us to find the most common value for alphabetical columns...The result of this is used to fill missing values
fuel_unit_mode = df['fuel_unit'].mode()[0]
fuel_unit_mode
#Replacing the missing values in the fuel_unit column with the mode.
#The below helps us select a specific column in our data-set and replace the missing value....Here since we want to replace the missing value with the mode which we have already gotten above we will input the variable used to store the mode which is fuel_unit_mode
df['fuel_unit'].fillna(fuel_unit_mode, inplace=True)
```

```python
#Question 8
corelation = df.corr()
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns,annot=True)
```

```python
#Question 9
total_fuel_burned_98 = df['fuel_cost_per_unit_burned'].iloc[1998].sum() 
total_fuel_burned_94 = df['fuel_cost_per_unit_burned'].iloc[1994].sum()
difference = total_fuel_burned_98 - total_fuel_burned_94
percentage_change = (difference/total_fuel_burned_98) * 100
percentage_change
```
